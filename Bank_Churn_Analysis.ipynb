```python
# find state(s) with highest number of churned customers and churn rate by state
mask_churned = df['Customer_Status'].astype(str).str.strip().str.lower() == 'churned'
churn_counts = df[mask_churned]['State'].value_counts()
total_by_state = df['State'].value_counts()

if churn_counts.empty:
    print("No churned customers found in the dataset.")
else:
    # summary dataframe
    state_churn_df = pd.DataFrame({
        'Churn_Count': churn_counts,
        'Total_Customers': total_by_state
    }).fillna(0).astype({'Churn_Count': int, 'Total_Customers': int})
    state_churn_df['Churn_Rate_%'] = state_churn_df['Churn_Count'] / state_churn_df['Total_Customers'] * 100

    # state(s) with highest churn count
    max_count = state_churn_df['Churn_Count'].max()
    top_count_states = state_churn_df[state_churn_df['Churn_Count'] == max_count]
    print("State(s) with highest churn count:")
    print(top_count_states[['Churn_Count', 'Total_Customers', 'Churn_Rate_%']].to_string())

    # state(s) with highest churn rate (require at least 10 customers to avoid tiny-sample noise)
    min_customers = 10
    eligible = state_churn_df[state_churn_df['Total_Customers'] >= min_customers]
    if not eligible.empty:
        max_rate = eligible['Churn_Rate_%'].max()
        top_rate_states = eligible[eligible['Churn_Rate_%'] == max_rate]
        print("\nState(s) with highest churn rate (min customers = %d):" % min_customers)
        print(top_rate_states[['Churn_Count', 'Total_Customers', 'Churn_Rate_%']].to_string())
    else:
        print("\nNo state has at least %d customers to compute a reliable churn rate." % min_customers)
```

```python

# Cell 14: Stacked normalized bar chart: distribution of top churn reasons across top 5 states

# Count churn per state
churn_counts = df[df['Churned'] == 1]['State'].value_counts()

# Top 5 churn-heavy states
top_states = churn_counts.head(5).index.tolist()

# Top 6 churn reasons
churn_reason_df = df[df['Churned'] == 1]['Churn_Reason'].value_counts().reset_index()
churn_reason_df.columns = ['Churn_Reason', 'Count']
top_reasons = churn_reason_df['Churn_Reason'].head(6).tolist()

# Filter data for top states
df_top = df[df['State'].isin(top_states)].copy()

if not df_top.empty:
    # Pivot table for state vs churn reason
    pivot = pd.crosstab(df_top['State'], df_top['Churn_Reason'])
    
    # Ensure all top reasons are in columns
    for r in top_reasons:
        if r not in pivot.columns:
            pivot[r] = 0

    pivot = pivot[top_reasons]
    pivot_norm = pivot.div(pivot.sum(axis=1), axis=0).fillna(0)

    colors = sns.color_palette("tab10", n_colors=len(top_reasons))
    fig, ax = plt.subplots(figsize=(10, 6))
    left = np.zeros(len(pivot_norm))
    states = pivot_norm.index.tolist()

    for i, reason in enumerate(top_reasons):
        vals = pivot_norm[reason].values
        ax.bar(states, vals, bottom=left, color=colors[i], label=reason)

        # Annotate large segments
        for j, v in enumerate(vals):
            if v > 0.05:
                ax.text(j, left[j] + v/2, f"{v*100:.0f}%", ha='center', va='center', color='white', fontsize=9)
        left += vals

    ax.set_ylabel("Proportion of Churn Reasons")
    ax.set_title("Top Churn Reasons within Top 5 States (Normalized)", fontsize=14)
    ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title="Churn Reasons")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("No data for top states to plot stacked distributions.")
```

```python
# Cell 15: Scatter plot - Balance vs Estimated Salary colored by Churn Reason

if {'Balance', 'EstimatedSalary', 'Churn_Reason'}.issubset(df.columns):
    # Keep top 6 churn reasons and group others as "Other"
    top_n = 6
    churn_reason_df = df[df['Churned']==1]['Churn_Reason'].value_counts().reset_index()
    churn_reason_df.columns = ['Churn_Reason', 'Count']
    top_reasons = churn_reason_df['Churn_Reason'].head(top_n).tolist()

    df_scatter = df.copy()
    df_scatter['Reason_Short'] = df_scatter['Churn_Reason'].where(df_scatter['Churn_Reason'].isin(top_reasons), 'Other')

    plt.figure(figsize=(9,6))
    sns.scatterplot(
        data=df_scatter[df_scatter['Churned']==1],
        x='Balance', y='EstimatedSalary',
        hue='Reason_Short',
        palette=sns.color_palette('tab10', n_colors=len(set(df_scatter['Reason_Short']))),
        alpha=0.8, s=60
    )
    plt.title("Balance vs Estimated Salary (Colored by Churn Reason)", fontsize=14)
    plt.xlabel("Balance (‚Çπ)")
    plt.ylabel("Estimated Salary (‚Çπ)")
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title='Churn Reason')
    plt.tight_layout()
    plt.show()
else:
    print("Required columns for scatter plot missing.")
```
